# **Date Conversions**

```python
rat_requests['created_date'] = pd.to_datetime(rat_requests['created_date'])
```

```python
rat_requests['closed_date'] = pd.to_datetime(rat_requests['closed_date'])
```

```python
rat_requests.info()
```

```python
rat_requests['created_date'] = pd.to_datetime(rat_requests['created_date'])
```

```python
rat_requests['created_year'] = rat_requests['created_date'].dt.year
```

```python
rat_requests['created_year'].value_counts()
```

```python
rat_requests['created_month'] = rat_requests['created_date'].dt.month
```

```python
rat_requests['created_month'].value_counts()
```

```python
rat_requests['created_day'] = rat_requests['created_date'].dt.day
```

```python
rat_requests['created_day'].value_counts()
```

```python
rat_requests.head()
```

```python
rat_requests['count_row'] = 1
```

```python
rat_requests['location_type'].value_counts()
```

```python
residential_list = ['3+ Family Apt. Building', '1-2 Family Dwelling' , '3+ Family Mixed Use Building', '1-2 Family Mixed Use Building', '3+ Family Apartment Building', 'Single Room Occupancy (SRO)', 'Residential Building', 'Apartment', 'Residence', '1-3 Family Dwelling', '3+ Family Apt.', 'Private Residence', '1-2 FamilyDwelling', '3+ Family Apt', '3+ Family']
```

```python
len(residential_list)
```

```python
rat_requests['is_residential'] = np.where(rat_requests['location_type'].isin(residential_list)==True, 'is residential', 'not residential')
```

```python
rat_requests['is_residential'].value_counts()
```

```python
rat_requests.groupby(['borough', 'is_residential'])['count_row'].sum().reset_index()
```

```python
rat_requests.groupby(['is_residential'])['count_row'].sum().reset_index()
```

```python
rat_request_res = rat_requests[rat_requests['is_residential'] == 'is residential']
```

```python
rat_request_nonres = rat_requests[rat_requests['is_residential'] == 'not residential']
```

```python
rat_request_nonres.shape
```

```python
rat_request_res.groupby(['borough', 'created_month'])['count_row'].sum().reset_index()
```

```python
rat_requests['tot_row'] = sum(rat_requests['count_row'])
```

```python
 conditions = [(rat_requests['created_month']== 6) | (rat_requests['created_month']== 7) | (rat_requests['created_month']== 8),
              (rat_requests['created_month']== 9) | (rat_requests['created_month']== 10) | (rat_requests['created_month']== 11),
              (rat_requests['created_month']== 12) | (rat_requests['created_month']== 1) | (rat_requests['created_month']== 2),
              (rat_requests['created_month']== 3) | (rat_requests['created_month']== 4) | (rat_requests['created_month']== 5)]
choices = ['summer', 'fall', 'winter', 'spring']
rat_requests['season'] = np.select(conditions, choices)
```

```python
rat_requests['season'].value_counts()
```

```python
rat_requests['duration'] = rat_requests['closed_date'] - rat_requests['created_date']
```

```python
rat_requests['duration'] = rat_requests['duration'].dt.days
```

```python
rat_requests.sample(3)
```

```python
rat_requests.groupby(['borough', 'season'])['count_row'].sum().reset_index()
```

```python
rat_requests.groupby(['borough', 'season'])['count_row'].max().reset_index()
```
