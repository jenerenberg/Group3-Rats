# **Import Libraries**



```python
import pandas as pd
import numpy as np
import requests
import geopandas as gpd
```

# **Data Imported**

```python
# query NYC 311 for complaint types with Rodent in the title
query = "https://data.cityofnewyork.us/resource/erm2-nwe9.csv?$query=" +  requests.utils.quote("SELECT * WHERE complaint_type = 'Rodent' LIMIT 1000000")
```

```python
rat_requests  = pd.read_csv(query)
```

# **Date Conversions**

```python
rat_requests['created_date'] = pd.to_datetime(rat_requests['created_date'])
```

```python
rat_requests['closed_date'] = pd.to_datetime(rat_requests['closed_date'])
```

```python
rat_requests.info()
```

**Created year, month, and day follow tidy data principles through adding variable names (year, day, month) and assigning values to them in each row. Every observation has a value with these variables.**

```python
rat_requests['created_year'] = rat_requests['created_date'].dt.year
```

```python
rat_requests['created_month'] = rat_requests['created_date'].dt.month
```

```python
rat_requests['created_month'].value_counts()
```

```python
rat_requests['created_year'].value_counts()
```

```python
rat_requests['created_day'] = rat_requests['created_date'].dt.day
```

# **Measurement Of Months & Seasons**

**Creating a new variable, seasons, with values summer, spring, winter, fall, to measure which had the most rat requests.**

**Using conditionals.**

```python
conditions = [(rat_requests['created_month']== 6) | (rat_requests['created_month']== 7) | (rat_requests['created_month']== 8),
              (rat_requests['created_month']== 9) | (rat_requests['created_month']== 10) | (rat_requests['created_month']== 11),
              (rat_requests['created_month']== 12) | (rat_requests['created_month']== 1) | (rat_requests['created_month']== 2),
              (rat_requests['created_month']== 3) | (rat_requests['created_month']== 4) | (rat_requests['created_month']== 5)]
choices = ['summer', 'fall', 'winter', 'spring']
rat_requests['season'] = np.select(conditions, choices)
```

```python
rat_requests['season'].value_counts()
```

```python
season_requests = rat_requests['season'].value_counts().reset_index()
```

```python
season_requests.head()
```

```python
rat_requests['duration'] = rat_requests['closed_date'] - rat_requests['created_date']
```

```python
rat_requests['duration'] = rat_requests['duration'].dt.days
```

```python
rat_requests['duration'].describe()
```

```python
rat_requests.groupby(['borough', 'season'])['count_row'].sum().reset_index()
```

**Requests by month (adding month names so they show on the chart)**

```python
rat_request_month['month'] = rat_request_month['created_month'].map(month_map)
```

```python
rat_request_month_txt = rat_request_month.drop('created_month', axis=1)
```
